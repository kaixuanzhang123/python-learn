# Series对象的创建

```python
import pandas as pd

# 直接通过列表创建Series
s = pd.Series([1,2,3,4])
# print(type(s))
# 通过列表创建Series时指定索引
s = pd.Series(data = [1,2,3,4],index = ['a','b','c','d'])
# 通过列表创建Series时指定索引和名称
s = pd.Series([1,2,3,4],index = ['a','b','c','d'],name = 'atguigu')

# 通过字典创建Series对象
dict = {"a":1,"b":2,"c":3,"d":4}
s = pd.Series(dict,name = 'atguigu')
s = pd.Series(dict,index=["a","d"],name = 'atguigu')
s
```

# Bool索引

```python
s = pd.Series({"a": -1.2, "b": 3.5, "c": 6.8, "d": 2.9})
bools = s > s.mean()
print(type(bools))
s[bools]
```
