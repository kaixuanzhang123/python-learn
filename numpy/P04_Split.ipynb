# 该案例演示了ndarray的切片和索引

```python
import numpy as np
arr = np.arange(10)
print(arr)

#获取索引为2的数据
print(arr[2])
# 从索引2开始到索引9(不包含)停止，间隔为2
print(arr[2:9:2])
# 从索引2开始到索引9(不包含)停止，间隔为2
print(arr[slice(2, 9, 2)])
# 从索引2开始到最后(不包含)，默认间隔为1
print(arr[2:])
# 从索引2开始到索引9(不包含)结束，默认间隔为1
print(arr[2:9])
```
