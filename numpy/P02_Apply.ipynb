```python
# 在Series上应用apply
import pandas as pd

# 定义一个对Series的处理函数  当前函数接收的参数就是Series中的一个元素
def func(item):
    if item == 10:
        return item * 10
    elif item == 20:
        return item * 5
    else:
        return item * 2

s = pd.Series([10, 20, 30])
# print(s * 10)
# s.apply(func)

# 可以给apply传递lambda表达式   lambda x:
# s.apply(lambda x: x*5)

# 传递带参数的函数  item代表series中的元素，parm1是函数执行的时候，需要的参数
def func(item,parm1):
    return item * parm1

s.apply(func,parm1 = 2)


```

```python
# DataFrame中使用apply
# 定义一个处理函数  接收的参数是一个Series对象
def func(s):
    return s.sum()

df = pd.DataFrame({"a": [10, 20, 30], "b": [40, 50, 60]})
# df
df.apply(func,axis = 1)
```

```python
# 问题说明
def func(s):
    return s["a"] / s["b"]

df = pd.DataFrame({"a": [10, 20, 30], "b": [40, 50, 60]})
df.apply(func,axis = 1)
```
