# Matplotlib可视化

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建画布，并指定画布大
plt.figure(figsize=(10, 6))
# 创建2行1列个子图，并指定第1个子图
plt.subplot(2,1,1)
# 设置x轴范围
plt.xlim(0,10)
# 设置y轴范围
plt.ylim(-1,1)
# 设置x轴的标签
plt.xlabel('x')
# 设置y轴的标签
plt.ylabel('sin(x)')
plt.title('sin(x)')
plt.plot(x, y1)  # 绘制曲线


# 创建2行1列个子图，并指定第1个子图
plt.subplot(2,1,2)
# 设置x轴范围
plt.xlim(0,10)
# 设置y轴范围
plt.ylim(-1,1)
# 设置x轴的标签
plt.xlabel('x')
# 设置y轴的标签
plt.ylabel('cos(x)')
plt.title('cos(x)')
plt.plot(x, y2)  # 绘制曲线


```
