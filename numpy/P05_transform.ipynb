# 分组转换

```python
import pandas as pd
import numpy as np
df = pd.read_csv("E:\\大模型课程\\05_尚硅谷大模型技术之Numpy&Pandas\\2.资料\\data\\employees.csv")
# df.groupby("department_id")["salary"].mean()
# df["部门平均薪水"] = df.groupby("department_id")["salary"].transform("mean")
# df

na_index = pd.Series(df.index.tolist()).sample(30)  # 随机挑选30条数据
na_index

df.loc[na_index, "salary"] = pd.NA  # 将这30条数据的salary设置为缺失值
df
# df.groupby("department_id")["salary"].agg(["size", "count"]) # 查看每组数据总数与非空数据数

def func(x):
    # 使用平均值填充，如果平均值也为NaN，用0填充
    if np.isnan(x.mean()):
        return 0
    return x.fillna(x.mean())

df.groupby("department_id")["salary"].transform(func)
# df.groupby("department_id")["salary"].agg(["size", "count"])
df.groupby("department_id").groups

```
