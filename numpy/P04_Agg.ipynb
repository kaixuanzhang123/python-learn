# 分组聚合

```python
import pandas as pd
df = pd.read_csv("E:\\大模型课程\\05_尚硅谷大模型技术之Numpy&Pandas\\2.资料\\data\\employees.csv")
# 按照部门id进行分组，统计出每个部门的最低薪水
# df.groupby("department_id")["salary"].min()
# df.groupby("department_id")["salary"].max()
# df.groupby("department_id")["salary"].median()
# df.groupby("department_id")["salary"].agg(["min","max","median"])

df.groupby("department_id").agg({"job_id": "nunique", "commission_pct": "mean","salary": "sum"}).rename(columns={"job_id": "工作岗位数", "commission_pct": "佣金均值","salary": "薪水和"})
```

```python
# 自定义函数完成聚合逻辑
"""统计每个部门员工last_name的首字母"""
def func(s):
    res = set()
    for i in s:
        res.add(i[0])
    return res


df = pd.read_csv("E:\\大模型课程\\05_尚硅谷大模型技术之Numpy&Pandas\\2.资料\\data\\employees.csv")
df.groupby("department_id")["last_name"].agg(func)
```
